```python
import csv
import itertools
import pandas as pd
import numpy as np
import networkx as nx
import pickle
```

```python
#Opens training labels ob file
file_to_read= open('training_labels_combined.ob', 'rb')
y_train = pickle.load(file_to_read)

#Opens training_set
X_train=pd.read_csv("train_transformed.csv")
X_train.drop(['Unnamed: 0'],axis=1, inplace=True)
X_train.drop(['count'],axis=1, inplace=True)

X_train.rename(columns={'Author_A':'Source'}, inplace=True)
X_train.rename(columns={'Author_B':'Sink'}, inplace=True)
#Opens test_set
X_test=pd.read_csv("test_transformed.csv")
X_test.drop(['Unnamed: 0'],axis=1, inplace=True)
X_test.drop(['Id'],axis=1, inplace=True)
X_train
```

```python

X_test
```

```python
import xgboost as xgb

xgb_model = xgb.XGBRegressor(objective="binary:logistic")

xgb_model.fit(X_train, y_train)
```

```python
result=pd.DataFrame(xgb_model.predict(X_test))
result.reset_index(inplace=True)
result.columns=['Id','Predicted']
result['Id']=[i+1 for i in result['Id']]
result.to_csv('preds.csv',index=False)
result
```
