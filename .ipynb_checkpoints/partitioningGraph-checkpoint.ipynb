```python
import pandas as pd
import networkx as nx
```

```python
df = pd.read_csv("dataset_pos.csv")
ng = nx.convert_matrix.from_pandas_edgelist(df, source= 'Author_A', target = 'Author_B', edge_attr= True)
df
```

```python
df.sort_values('Author_A',inplace=True, ascending=True)
```

```python
df_train_pos = df.iloc[:3015,:]
```

```python
df_test_pos = df.iloc[3016:,:]
```

```python
df_neg = pd.read_csv("dataset_neg.csv")
df_neg.sort_values('Author_A',inplace=True, ascending=True)
```

```python
df_train_neg = df_neg.iloc[:3012,:]
```

```python
df_test_neg = df_neg.iloc[3013:,:]
```

```python
pd.concat([df_train_pos, df_train_neg], axis = 0).to_csv('df_train_partition.csv')
```

```python
pd.concat([df_test_pos, df_test_neg], axis = 0).to_csv('df_test_partition.csv')
```
